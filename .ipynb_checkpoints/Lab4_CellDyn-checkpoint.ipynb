Task 1 Begin

```python
kp = 0.1
km = 0.3
random.seed(3)

To = 1/km
Tc = 1/kp

r = random.random()
start=0.5
if r<0.5:
    start=0
else:
    start=1

current = start
dt = 1
for t in range(100):
    Q = np.array([[1-(kp*dt), (km*dt)],[(kp*dt), 1-(km*dt)]])
    if current==0:
        y = random.random()
        if y<Q[0][1]:
            current = 1
            print("Open->Closed@ "+str(t))
            dt = 0
    else:
        x = random.random()
        if x<Q[1][0]:
            current = 0
            print("Closed->Open@ "+str(t))
            dt = 0
    dt += 1
```

```python
currents = start
t = 1
while t<=100:
    rtwo = random.random()
    TC = (1/kp) *np.log(1/rtwo)
    TO = (1/km) *np.log(1/rtwo)
    if currents==0:
        print("op>clo @ "+str(t+TO))
        t += TO
        currents = 1
    else:
        print("clo>op @ "+str(t+TC))
        t += TC
        currents = 0
```

Task 2 Begin
-------

```python
Mextra = 1.2 #extracellular
Mmax = 1000
M = 0
km = 5 #ratecon
t = 0
Mt = 0 #startVal
kpl = 0.1 #k+
kmi = 0.3 #k-
a1 = kpl #propensity open
a2 = kmi #propensity close
a3 = km*Mextra #propensity pass ion >>> 0 is closed
currents = 1
ooo = []
ccc = []
ppp = []
ttt = []
while M<Mmax:
    r1 = random.random()
    r2 = random.random()
    if currents == 0:
        a0 = a1+a2+a3
    else:
        a0 = a1+a2
    tau = (1/a0)*np.log(1/r1)
    if currents == 0:
        if r2>=(1/a0)*a1 and r2<(1/a0)*(a1+a2):
            #print("reaction closing")
            ccc.append([t,M])
            currents = 1
        elif r2>=(1/a0)*(a1+a2):
            #print("reaction ion passing")
            M = M+1
            ppp.append([t,M])
    else:
        if r2<(1/a0)*a1:
            #print("reaction opening")
            ooo.append([t,M])
            currents = 0
    ttt.append([t,M])
    t=t+1
#print(ppp[:500])
```

```python
ppp = np.asarray(ppp)
ccc = np.asarray(ccc)
ooo = np.asarray(ooo)
ttt = np.asarray(ttt)
fivehundo = ppp[:500,0]
comp_open = []
for i in range(len(ccc)):
    for j in range(int(ooo[i][0]),int(ccc[i][0]+1)):
        comp_open.append([j,ttt[j][1]])
comp_closed = []
for i in range(len(ooo)-1):
    for j in range(int(ccc[i][0]),int(ooo[i+1][0])):
        comp_closed.append([j,ttt[j][1]])
#print(comp_closed)
comp_open = np.asarray(comp_open)
comp_closed = np.asarray(comp_closed)
```

```python
fig = plt.figure(figsize=(20,20))
maxT = fivehundo[499]
for i in range(int(maxT)):
    if i in comp_closed[:,0]:
        c = 'r'
    elif i in fivehundo[:]:
        c = 'g'
    else:
        c = 'b'
    plt.scatter(ttt[i,0],ttt[i,1],color=c)
    #blue is open (but not passing)
    #green is open and passing
    #red is closed (and not passing)
```

```python
#D in cm2/sec
D = 0.000005

e = np.random.normal(0,1,2)
print(e)

X = np.zeros(500)
Y = np.zeros(500)

X[0] = np.sqrt(2*D)*e[0]
Y[0] = abs(np.sqrt(2*D)*e[1])
e = np.random.normal(0,1,2)
X[1] = X[0] + np.sqrt(2*D)*e[0]
Y[1] = abs(Y[0] + np.sqrt(2*D)*e[1])

fig = plt.figure(figsize = (15,15))

for t in range(1,499):
    e = np.random.normal(0,1,2)
    dt = fivehundo[t]-fivehundo[t-1]
    X[t+dt] = X[t] + np.sqrt(2*D*dt)*e[0]
    Y[t+dt] = abs(Y[t] + np.sqrt(2*D*dt)*e[1])
plt.scatter(X,Y)
plt.plot([np.average(X),np.average(X)],[0,max(Y)], c='pink')
plt.plot([min(X),max(X)],[0,0], c='cyan')
    

#cell boundaries in micrometres
yb = 0
yt = 100
xl = -50
xr = 50
```
